# Overview

Are there any predictable patterns that determine churn for the Telecom Company? This analysis seeks to classify customer who leave the service (churn) and determine which levers can be pulled to increase customer retention.

# Business Problem


Customers who leave Telecom's service quickly result in money lost on the customer acquisition and other related costs. This analysis will classify customers who churn and interpret what leads to their departure.

# Data Preparation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.metrics import classification_report
%matplotlib inline
```

Significant class imbalance; manipulation needed for modeling

```python
df1 = df.drop(['phone number'], axis=1) #drop phone number, not relevant
```

```python
df1.head()
```

```python
df1.isna().sum() #check for N/A
```

```python
#Need to transform categorical variables:

total_col_names = df1.columns
# find numeric columns 
num_cols = df1._get_numeric_data().columns
# getting category columns 
cat_col_names = list(set(total_col_names) - set(num_cols))
```

```python
def label_encoding(col_name):
  le = LabelEncoder()
  df1[col_name] = le.fit_transform(df1[col_name])
    
for col_name in cat_col_names:
  label_encoding(col_name)

df1.head()
```

```python
label_encoding('churn')
```

```python
df1.head()
```

```python
y = df1['churn'] #build baseline model 
X = df1.drop(['churn'], axis=1)

X_train, y_train, X_test, y_test = train_test_split(X, y, random_state=42)
```

```python
scaler = StandardScaler()

scaled_data_train = scaler.fit_transform(X_train)
scaled_data_test = scaler.fit(y_test)

scaled_df1 = pd.DataFrame(scaled_data_train, columns=df1.columns)
scaled_df1.head()
```

```python
#create pipeline for scaling and KNN
pipe_scaler_KNN = Pipeline([('ss', StandardScaler()),
                             ('knn', KNeighborsClassifier())])
```

```python
# Fit the training data to pipeline
pipe_scaler_KNN.fit(X_train, y_train)

# Print the accuracy on test set
pipe_scaler_KNN.score(X_test, y_test)
```
